## 2. Nettoyage des données

```python
# Remplacer -1 par NaN dans pdays
df['pdays'] = df['pdays'].replace(-1, np.nan)
df['pdays']
```

```python
df['no_previous_contact'] = df['pdays'].isna().astype(int)
df['no_previous_contact']
```

```python
df['pdays'] = df['pdays'].fillna(0)
df['pdays']
```

```python
#  Aperçu unique par colonnes ---
unique_counts = df.nunique() 
unique_counts 

```

```python
# Analyse univariée (histogrammes & stats)
import matplotlib.pyplot as plt
import seaborn as sns

# Variables numériques ---
numeric_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']

# Histograms for numerical variables
fig, axes = plt.subplots(len(numeric_cols), 1, figsize=(8, 20))
for i, col in enumerate(numeric_cols):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f"Distribution de {col}")
plt.tight_layout()
plt.show()

# Variables catégorielles ---
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
categorical_counts = {col: df[col].value_counts() for col in categorical_cols}

# Taux de 'yes' dans la variable cible y ---
target_rate = df['y'].value_counts(normalize=True) * 100
categorical_counts, target_rate
```

Analyse bivariée

```python
# C’est l’étude de la relation entre une variable explicative (par exemple l’âge, le métier, la durée de l’appel) et la variable cible (y = yes/no).
# La corrélation mesure la force de la relation linéaire entre deux variables.
# Pour savoir si l’âge, la durée ou le solde (balance) influencent y, on convertit y en 0 (no) et 1 (yes) :

df['y_binary'] = df['y'].map({'yes': 1, 'no': 0})
df['y_binary']


```

```python
# Ensuite, on calcule la corrélation avec les variables numériques :
df.corr(numeric_only=True)['y_binary']
# Une corrélation > 0.2 (ou < -0.2) peut indiquer une relation intéressante.

```

```python
# (b) Taux de souscription par catégorie
df.groupby('job')['y_binary'].mean() * 100 #: resultat = pourcentage de "yes" par type de métier.


# Pour les variables catégorielles comme job, marital, education 
```

```python
# Visualisations
# Barplots pour les variables catégorielles :
sns.barplot(x='job', y='y_binary', data=df)

```

```python
# OU
sns.boxplot(x='y', y='age', data=df)
```

##  Interprétation typique du dataset

#### Duration : plus l’appel est long, plus la probabilité de dire 'yes' augmente (corrélation souvent > 0.4).
#### Age : les jeunes (<30 ans) et les très vieux (>60 ans) souscrivent moins.
#### Job : les "retired" (retraités) ou "students" souscrivent plus souvent.
#### Housing/Loan : ceux qui n’ont pas de prêt immobilier souscrivent souvent plus.



##  Conclusion et Insights possibles

#### La variable duration est généralement la plus corrélée avec y.
#### Les clients retraités ou étudiants semblent plus ouverts à souscrire.
#### Il existe un déséquilibre de la cible (88 % de no), donc il faudra équilibrer les classes s'il faut  construire un modèle prédictif.
